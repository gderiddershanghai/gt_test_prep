## Q1 Solution
Observations (3 batches of size 4):

- Batch 1 mean:  
$$
\frac{140 + 155 + 160 + 148}{4} = \frac{603}{4} = 150.75
$$

- Batch 2 mean:  
$$
\frac{170 + 165 + 150 + 172}{4} = \frac{657}{4} = 164.25
$$

- Batch 3 mean:  
$$
\frac{180 + 175 + 160 + 155}{4} = \frac{670}{4} = 167.5
$$

Grand mean:
$$
\bar{X} = \frac{150.75 + 164.25 + 167.5}{3} = 160.83\overline{3}
$$

Batch sample variance:

First compute:
$$
\text{Deviations:}\quad (-10.08)^2, (3.42)^2, (6.67)^2
$$
where deviations are relative to 160.833.

So:
- \( (150.75-160.833)^2 \approx 101.71 \)
- \( (164.25-160.833)^2 \approx 11.68 \)
- \( (167.5-160.833)^2 \approx 44.44 \)

Sum of squared deviations:

$$
101.71 + 11.68 + 44.44 = 157.83
$$

Variance (sample, denominator = \( b-1 = 2 \)):

$$
S_B^2 = \frac{157.83}{2} = 78.915
$$

Now, 90% CI:

$$
\text{Standard error} = \sqrt{\frac{S_B^2}{n}} = \sqrt{\frac{78.915}{12}} \approx \sqrt{6.57625} \approx 2.564
$$

Critical value \( t_{0.05,2} \approx 2.92 \).

Thus, margin of error:

$$
2.92 \times 2.564 \approx 7.49
$$

Final CI:

$$
[160.83 - 7.49, 160.83 + 7.49] = [153.34, 168.32]
$$

✅

$$
\text{Var}(\bar{X}_3) = \frac{1}{3}\left(R_0 + 2\left(1-\frac{1}{3}\right)R_1 + 2\left(1-\frac{2}{3}\right)R_2\right)
$$

Compute inside:

- \( 2(2/3)(3.5) = 4.6667 \)
- \( 2(1/3)(3) = 2 \)

Sum:

$$
4 + 4.6667 + 2 = 10.6667
$$

Thus:

$$
\text{Var}(\bar{X}_3) = \frac{10.6667}{3} = 3.5556
$$
