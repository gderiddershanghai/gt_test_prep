### **Question 15**  
**Correct Answer:** C. Critical fractile based on overage and underage cost  
**Explanation:**  
The newsvendor model determines the optimal one-time order quantity based on balancing the **cost of overstocking (CO)** and **understocking (CU)**. This balance is captured in the **critical fractile**:  
$$
\text{Critical Fractile} = \frac{CU}{CU + CO}
$$  
This ratio directly determines the service level and optimal order quantity.

---

### **Question 16**  
**Correct Answer:** C. Shortages are allowed  
**Explanation:**  
The basic EOQ model **does not** allow shortages. It assumes:
- Constant demand  
- Instantaneous replenishment  
- No stockouts  
- Cost focus only on ordering and holding  
Allowing shortages would require a modified model.

---

### **Question 17**  
**Correct Answer:** B. EOQ decreases  
**Explanation:**  
The EOQ formula is:  
$$
EOQ = \sqrt{\frac{2DS}{H}}
$$  
If holding cost ($H$) increases, the denominator increases, causing the EOQ value to decrease. This means firms should order smaller quantities more frequently when it becomes more expensive to hold inventory.

---

### **Question 18**  
**Correct Answer:** B. Holding and ordering  
**Explanation:**  
EOQ is derived by minimizing the **sum** of:
- **Ordering costs** (cost incurred per order placed)  
- **Holding costs** (cost to carry inventory over time)  
Other costs like shrinkage, spoilage, or transportation are excluded in the basic EOQ model.

---

### **Question 19**  
**Correct Answer:** C. Because excess inventory is kept to avoid stockouts  
**Explanation:**  
Achieving a 100% fill rate requires **buffering against all demand uncertainty**, which leads to carrying large amounts of safety stock. This increases **holding costs** significantly and is typically not optimal in cost-sensitive environments.

---

### **Question 20**  
**Correct Answer:** A. It represents the lost margin on a missed sale  
**Explanation:**  
In a newsvendor setting, the **underage cost (CU)** is calculated as the **contribution margin lost** when demand exceeds inventory. This is approximated as:  
$$
CU = \text{Retail Price} - \text{Wholesale Cost}
$$  
It reflects the opportunity cost of failing to meet demand.

### Question 31  
**Correct Answer:** C. Variance ratio of orders to sales

$$

$$  
Higher values indicate greater distortion and volatility.
